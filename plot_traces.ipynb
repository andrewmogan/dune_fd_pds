```python
from os import listdir
import numpy as np
import matplotlib.pyplot as plt
```

```python
folderpath = '/Users/samuelfogarty/OneDrive/Desktop/OneDrive/photondetectortests/dune_fd_pds/data/June82022_simlabtests/'
#filename = 'wvfm_module4_topLED_60ns_3.6V_44.394Bias_000_ALL.csv'
filename = 'wvfm_25MHzSR_44.329VBias_10msPerDiv_2.5Mpts_module4_000_ALL.csv'
filepath = folderpath + filename 

trace = np.loadtxt(filepath,delimiter=',', skiprows=12, usecols=(0,1,2,3,4))
```

```python
t = trace[:,0][0:100000]
V1 = trace[:,1][0:100000]
V2 = trace[:,2][0:100000]
V3 = trace[:,3][0:100000]
V4 = trace[:,4][0:100000]
```

```python
fig, axes = plt.subplots(nrows=4, ncols=1, sharex=True, sharey=False, figsize=(10,8))
axes[0].plot(t*1e6, V1,'r-',label='CH1')
axes[1].plot(t*1e6, V2,'b-',label='CH2')
axes[2].plot(t*1e6, V3,'g-',label='CH3')
axes[3].plot(t*1e6, V4,'y-',label='CH4')
axes[3].set_xlabel('usec')
axes[0].set_ylabel('V')
axes[1].set_ylabel('V')
axes[2].set_ylabel('V')
axes[3].set_ylabel('V')
axes[0].legend()
axes[1].legend()
axes[2].legend()
axes[3].legend()
#axes[0].set_title('Module 4 All Channels')
#plt.savefig('module4_allchannels_traces.pdf')
plt.show()
```
